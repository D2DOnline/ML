```python
all_reviews = np.load('C:\\Users\\jitendrakumar_si\\Downloads\\all_reviews_df.npy', allow_pickle=True)
```

```python
import pandas as pd
all_reviews_df = pd.DataFrame(all_reviews, columns=['review', 'sentiment', 'words', 'ids'])
```

```python
all_reviews_df
```

```python
def create_lookup(words):
  lookup={}
  for i in range(len(words)):
    try:
      word= words[i].decode('utf-8')
    except:
      word= words[i].encode().decode('utf-8')
    lookup[word]=i 
  return lookup

vocab = []
for words in all_reviews_df['words']:
  for word in words:
    vocab.append(word)
    
vocab_set = set(vocab)
vocab_dic = create_lookup(list(vocab_set))
```

```python
vocab_dic
```

```python
EMBEDDING_DIM = 50
```

```python
embedding_matrix = np.zeros((len(vocab_set), EMBEDDING_DIM))
embedding_matrix.shape
```

```python
for word, i in vocab_dic.items():
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        try:
            embedding_matrix[i] = embedding_vector
        except:
            pass
```
